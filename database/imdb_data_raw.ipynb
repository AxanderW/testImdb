```python
import pandas as pd
import os
```

```python
data_folder = 'imdb_data'
data_files = []
for _, _, files in os.walk(data_folder, topdown=False):
   for name in files:
      data_files.append(os.path.join(data_folder, name))
```

```python
data_files
```

```python
data_frames = []
for d in data_files:
    if '.gz' in d:
        data_frames.append(pd.read_csv(d, delimiter='\t', low_memory=False))
```
